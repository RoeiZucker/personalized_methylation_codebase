```python
import os
import math
import yaml

RESULTS_ROOT = "/cs/usr/roeizucker/new_storage/jupyter_notebooks/Tom_Hope_Project/results"


def _yaml_files(path):
    if not os.path.isdir(path):
        return []
    return sorted(
        os.path.join(path, f)
        for f in os.listdir(path)
        if f.endswith(".yaml")
    )


def _safe_load_yaml(path):
    try:
        return yaml.safe_load(open(path))
    except Exception:
        return None


def _epoch_ckpts(model_dir):
    if not os.path.isdir(model_dir):
        return []
    return [x for x in os.listdir(model_dir) if x.startswith("epoch-")]


def _bar(done, total, width=28):
    if total <= 0:
        return "[" + ("-" * width) + "] 0.0%"
    frac = max(0.0, min(1.0, done / total))
    fill = int(round(frac * width))
    return "[" + ("#" * fill) + ("-" * (width - fill)) + f"] {frac * 100:5.1f}%"


def _extraction_status(cfg):
    if not cfg:
        return "not_started"

    task = cfg.get("task", {})
    paths = cfg.get("paths", {})
    task_type = task.get("type", "")

    if task_type == CPG_SEPERATING_SITES_TASK_NAME:
        out_path = paths.get("per_variant_data_path")
        return "done" if out_path and os.path.exists(out_path) else "not_started"

    train_path = paths.get("hf_dataset_train_path")
    test_path = paths.get("hf_dataset_test_path")
    test_size = float(task.get("test_size", 0.0) or 0.0)

    train_ok = bool(train_path and os.path.exists(train_path))
    if not train_ok:
        return "not_started"
    if test_size == 0:
        return "done"
    test_ok = bool(test_path and os.path.exists(test_path))
    return "done" if test_ok else "in_progress"


def _training_status(cfg, completed_model_dirs=None):
    if not cfg:
        return "not_started"
    out_dir = cfg.get("paths", {}).get("output_dir")
    expected = int(cfg.get("train", {}).get("num_train_epochs", 0) or 0)
    if expected <= 0:
        expected = 1

    if not out_dir or not os.path.isdir(out_dir):
        if completed_model_dirs and out_dir in completed_model_dirs:
            return "done_deleted"
        return "not_started"

    ckpts = _epoch_ckpts(out_dir)
    if len(ckpts) == 0:
        if completed_model_dirs and out_dir in completed_model_dirs:
            return "done_deleted"
        return "not_started"
    if len(ckpts) < expected:
        return "in_progress"
    return "done"


def _eval_result_exists(eval_cfg):
    task = (eval_cfg or {}).get("task", {})
    analysis_name = task.get("analysis_name")
    base_suffix = task.get("base_suffix")

    if analysis_name and base_suffix:
        p = os.path.join(RESULTS_ROOT, base_suffix, f"{analysis_name}_result.csv")
        return os.path.exists(p)

    if analysis_name:
        legacy_dir = os.path.join(RESULTS_ROOT, analysis_name)
        return (
            os.path.exists(os.path.join(legacy_dir, "result.csv"))
            or os.path.exists(os.path.join(legacy_dir, "resilt.csv"))
        )
    return False


def _build_eval_completion_index(eval_yaml_paths):
    completed = set()
    for yp in eval_yaml_paths:
        cfg = _safe_load_yaml(yp)
        if not cfg or not _eval_result_exists(cfg):
            continue
        for mp in (cfg.get("paths", {}).get("model_paths", []) or []):
            parent = os.path.dirname(mp)
            if not parent:
                continue
            # Only retrain outputs are auto-marked as safely completed/deleted.
            # Pretrain outputs must be managed manually.
            if "_retrain" not in os.path.basename(parent):
                continue
            completed.add(parent)
    return completed


def _build_expected_epochs_map(training_yaml_paths):
    expected = {}
    for yp in training_yaml_paths:
        cfg = _safe_load_yaml(yp)
        if not cfg:
            continue
        out_dir = cfg.get("paths", {}).get("output_dir")
        n = cfg.get("train", {}).get("num_train_epochs")
        if out_dir and n is not None:
            expected[out_dir] = int(n)
    return expected


def _eval_ready(eval_cfg, expected_epochs_by_output_dir):
    model_paths = (eval_cfg or {}).get("paths", {}).get("model_paths", []) or []
    if not model_paths:
        return False
    if not all(os.path.exists(p) for p in model_paths):
        return False

    retrain_dirs = {os.path.dirname(p) for p in model_paths if os.path.basename(p).startswith("epoch-")}
    # Require at least one retrain checkpoint path; pretrain-only eval configs should not run.
    retrain_dirs = {d for d in retrain_dirs if "_retrain" in os.path.basename(d)}
    if not retrain_dirs:
        return False
    for d in retrain_dirs:
        # Only enforce full-epoch completion on retrain outputs.
        # Pretrain dirs can be partially trained as long as referenced checkpoints exist.
        if "_retrain" not in os.path.basename(d):
            continue
        required = expected_epochs_by_output_dir.get(d)
        if required is None:
            return False
        if len(_epoch_ckpts(d)) < required:
            return False
    return True


def _eval_status(eval_cfg, expected_epochs_by_output_dir):
    if _eval_result_exists(eval_cfg):
        return "done"
    return "ready" if _eval_ready(eval_cfg, expected_epochs_by_output_dir) else "blocked"


def _project_dirs_from_config(config_path):
    cfg = yaml.safe_load(open(config_path))
    p = cfg["params"]
    base_suffix = p["project_suffix"]
    root = p["created_configs_path"]
    if not os.path.isdir(root):
        return []
    return sorted(
        os.path.join(root, d)
        for d in os.listdir(root)
        if os.path.isdir(os.path.join(root, d)) and d.startswith(base_suffix + "_lr_")
    )


def _count_status(statuses, keys):
    c = {k: 0 for k in keys}
    for s in statuses:
        if s in c:
            c[s] += 1
    return c


def show_project_progress(config_path):
    project_dirs = _project_dirs_from_config(config_path)
    if not project_dirs:
        print("No project directories found for this config.")
        return

    all_extraction = []
    all_training = []
    all_eval = []

    print("Project progress summary")
    print("config:", config_path)
    print("projects found:", len(project_dirs))
    print("=" * 90)

    for d in project_dirs:
        variability_extraction = os.path.join(d, "variability_extraction")
        pretrain_extraction = os.path.join(d, "pretrain_extraction")
        retrain_extraction = os.path.join(d, "retrain_extraction")

        pretrain_training = os.path.join(d, "pretrain_training")
        lora_pretrain_training = os.path.join(d, "lora_pretrain_training")
        retrain_training = os.path.join(d, "retrain_training")
        eval_configs = os.path.join(d, "eval_configs")

        ext_yamls = (
            _yaml_files(variability_extraction)
            + _yaml_files(pretrain_extraction)
            + _yaml_files(retrain_extraction)
        )
        train_yamls = (
            _yaml_files(pretrain_training)
            + _yaml_files(lora_pretrain_training)
            + _yaml_files(retrain_training)
        )
        eval_yamls = _yaml_files(eval_configs)

        ext_status = [_extraction_status(_safe_load_yaml(p)) for p in ext_yamls]

        expected_map = _build_expected_epochs_map(train_yamls)
        completed_model_dirs = _build_eval_completion_index(eval_yamls)
        train_status = [_training_status(_safe_load_yaml(p), completed_model_dirs=completed_model_dirs) for p in train_yamls]
        eval_status = [_eval_status(_safe_load_yaml(p), expected_map) for p in eval_yamls]

        all_extraction.extend(ext_status)
        all_training.extend(train_status)
        all_eval.extend(eval_status)

        ext_counts = _count_status(ext_status, ["done", "in_progress", "not_started"])
        tr_counts = _count_status(train_status, ["done", "done_deleted", "in_progress", "not_started"])
        ev_counts = _count_status(eval_status, ["done", "ready", "blocked"])

        ext_total = len(ext_status)
        tr_total = len(train_status)
        ev_total = len(eval_status)

        print(os.path.basename(d))
        print(f"  extraction {ext_counts['done']}/{ext_total} {_bar(ext_counts['done'], ext_total)}")
        print(f"    in_progress={ext_counts['in_progress']} not_started={ext_counts['not_started']}")
        print(f"  training   {tr_counts['done'] + tr_counts['done_deleted']}/{tr_total} {_bar(tr_counts['done'] + tr_counts['done_deleted'], tr_total)}")
        print(f"    done_deleted={tr_counts['done_deleted']} in_progress={tr_counts['in_progress']} not_started={tr_counts['not_started']}")
        print(f"  eval       {ev_counts['done']}/{ev_total} {_bar(ev_counts['done'], ev_total)}")
        print(f"    ready_to_run={ev_counts['ready']} blocked={ev_counts['blocked']}")
        print("-" * 90)

    ext_counts = _count_status(all_extraction, ["done", "in_progress", "not_started"])
    tr_counts = _count_status(all_training, ["done", "done_deleted", "in_progress", "not_started"])
    ev_counts = _count_status(all_eval, ["done", "ready", "blocked"])

    ext_total = len(all_extraction)
    tr_total = len(all_training)
    ev_total = len(all_eval)

    done_all = ext_counts["done"] + tr_counts["done"] + tr_counts["done_deleted"] + ev_counts["done"]
    total_all = ext_total + tr_total + ev_total

    print("OVERALL")
    print(f"  total done {done_all}/{total_all} {_bar(done_all, total_all, width=40)}")
    print(f"  extraction done={ext_counts['done']} in_progress={ext_counts['in_progress']} not_started={ext_counts['not_started']}")
    print(f"  training   done={tr_counts['done']} done_deleted={tr_counts['done_deleted']} in_progress={tr_counts['in_progress']} not_started={tr_counts['not_started']}")
    print(f"  eval       done={ev_counts['done']} ready_to_run={ev_counts['ready']} blocked={ev_counts['blocked']}")


# usage
config_path = "/cs/usr/roeizucker/new_storage/jupyter_notebooks/Tom_Hope_Project/refactored_code/configs/project_configs/_sixteen_shards_of_adonalsium_Heart_Cardiomyocyte.yaml"
show_project_progress(config_path)





```
