```python
import os
import json
import random
from dataclasses import dataclass
from typing import List, Dict, Any, Optional

import torch
import torch.nn as nn
from torch.utils.data import Dataset

from evo2 import Evo2
from transformers import TrainingArguments, Trainer


# -------------------------
# 1) Toy dataset (sequence -> per-token labels)
# -------------------------
def make_random_dna(n: int) -> str:
    return "".join(random.choice("ACGT") for _ in range(n))

def labels_is_G(seq: str) -> List[int]:
    # token label: 1 if G else 0
    return [1 if c == "G" else 0 for c in seq]

class DNATokenDataset(Dataset):
    def __init__(self, n_samples: int, seq_len: int, seed: int = 0):
        random.seed(seed)
        self.rows = []
        for _ in range(n_samples):
            s = make_random_dna(seq_len)
            self.rows.append({"sequence": s, "labels": labels_is_G(s)})

    def __len__(self):
        return len(self.rows)

    def __getitem__(self, idx):
        return self.rows[idx]


# -------------------------
# 2) Collator that uses Evo2 tokenizer
# -------------------------
@dataclass
class Evo2TokenCollator:
    evo2_tokenizer: Any
    ignore_index: int = -100

    def __call__(self, batch: List[Dict[str, Any]]) -> Dict[str, torch.Tensor]:
        # Keep it simple: assume fixed length in this demo (no padding).
        input_ids_list = []
        labels_list = []

        for ex in batch:
            seq = ex["sequence"]
            lab = ex["labels"]

            ids = self.evo2_tokenizer.tokenize(seq)

            if len(ids) != len(lab):
                raise ValueError(
                    f"Token/label length mismatch: tokens={len(ids)} labels={len(lab)}. "
                    "If this happens in your real data, you must define a token<->base alignment rule."
                )

            input_ids_list.append(torch.tensor(ids, dtype=torch.int))
            labels_list.append(torch.tensor(lab, dtype=torch.long))

        input_ids = torch.stack(input_ids_list, dim=0)  # [B, L]
        labels = torch.stack(labels_list, dim=0)        # [B, L]

        return {"input_ids": input_ids, "labels": labels}


# -------------------------
# 3) Trainer-compatible wrapper module
# -------------------------
class Evo2TokenClassifier(nn.Module):
    """
    Head-only token classification on top of Evo2 embeddings.
    Evo2 backbone is used as a frozen feature extractor in this minimal example.
    """
    def __init__(
        self,
        evo2_model: Evo2,
        num_labels: int,
        layer_name: str = "blocks.28.mlp.l3",
        freeze_backbone: bool = True,
        ignore_index: int = -100,
    ):
        super().__init__()
        self.evo2 = evo2_model
        self.layer_name = layer_name
        self.freeze_backbone = freeze_backbone
        self.ignore_index = ignore_index

        # LazyLinear infers hidden_dim on first forward pass
        self.classifier = nn.LazyLinear(num_labels)
        self.loss_fct = nn.CrossEntropyLoss(ignore_index=ignore_index)

    def forward(self, input_ids: torch.Tensor, labels: Optional[torch.Tensor] = None):
        # Get per-token embeddings from Evo2
        if self.freeze_backbone:
            with torch.no_grad():
                _, embs = self.evo2(
                    input_ids,
                    return_embeddings=True,
                    layer_names=[self.layer_name],
                )
        else:
            # WARNING: this may or may not work efficiently with the Evo2 inference stack.
            _, embs = self.evo2(
                input_ids,
                return_embeddings=True,
                layer_names=[self.layer_name],
            )

        h = embs[self.layer_name]  # [B, L, H]
        logits = self.classifier(h)  # [B, L, C]

        out = {"logits": logits}

        if labels is not None:
            # Flatten for CE: (B*L, C) vs (B*L,)
            loss = self.loss_fct(
                logits.reshape(-1, logits.size(-1)),
                labels.reshape(-1),
            )
            out["loss"] = loss

        return out

    def save_pretrained(self, output_dir: str):
        os.makedirs(output_dir, exist_ok=True)
        torch.save(self.state_dict(), os.path.join(output_dir, "head_and_wrapper.pt"))
        meta = {
            "layer_name": self.layer_name,
            "freeze_backbone": self.freeze_backbone,
        }
        with open(os.path.join(output_dir, "meta.json"), "w") as f:
            json.dump(meta, f, indent=2)


# -------------------------
# 4) Metrics (token accuracy ignoring -100)
# -------------------------
def compute_metrics(eval_pred):
    # eval_pred.predictions: [N, L, C] ; label_ids: [N, L]
    preds = eval_pred.predictions.argmax(-1)
    labels = eval_pred.label_ids
    mask = labels != -100
    acc = (preds[mask] == labels[mask]).mean() if mask.any() else 0.0
    return {"token_acc": float(acc)}


# -------------------------
# 5) Run a tiny training
# -------------------------
device = "cuda:0" if torch.cuda.is_available() else "cpu"

evo2_backbone = Evo2("evo2_7b_262k")  # your checkpoint
# Evo2 splits across GPUs automatically when needed; input_ids are typically placed on cuda:0 in examples. :contentReference[oaicite:3]{index=3}

train_ds = DNATokenDataset(n_samples=128, seq_len=256, seed=0)
eval_ds  = DNATokenDataset(n_samples=32,  seq_len=256, seed=1)

collator = Evo2TokenCollator(evo2_backbone.tokenizer)

model = Evo2TokenClassifier(
    evo2_model=evo2_backbone,
    num_labels=2,
    layer_name="blocks.28.mlp.l3",   # from Evo2 README examples :contentReference[oaicite:4]{index=4}
    freeze_backbone=True,
).to(device)

args = TrainingArguments(
    output_dir="evo2_tokenclf_demo",
    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,
    gradient_accumulation_steps=8,
    num_train_epochs=1,
    learning_rate=1e-3,              # head-only: can be relatively high
    logging_steps=10,
    eval_strategy="steps",
    eval_steps=50,
    save_strategy="steps",
    save_steps=50,
    report_to="none",
)

# ---- initialize LazyLinear by running one batch through the model ----
batch = collator([train_ds[0]])  # one example
batch = {k: v.to(device) for k, v in batch.items()}

model.eval()
with torch.no_grad():
    _ = model(**batch)  # initializes the LazyLinear parameters
model.train()
# ---------------------------------------------------------------------


trainer = Trainer(
    model=model,
    args=args,
    train_dataset=train_ds,
    eval_dataset=eval_ds,
    data_collator=collator,
    compute_metrics=compute_metrics,
)

trainer.train()
trainer.save_model()  # calls model.save_pretrained()
```
