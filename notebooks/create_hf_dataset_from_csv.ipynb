```python
from transformers import AutoTokenizer
NUCLEPTIDE_TRANSFORMER_500M_1000G_NAME = "InstaDeepAI/nucleotide-transformer-500m-1000g"
NUCLEPTIDE_TRANSFORMER_2B_MS_NAME = "InstaDeepAI/nucleotide-transformer-2.5b-multi-species"
EVO2_40B_NAME = "arcinstitute/evo2_40b"
MODEL_NAME = NUCLEPTIDE_TRANSFORMER_500M_1000G_NAME  # or NUCLEPTIDE_TRANSFORMER_2B_MS_NAME or EVO2_40B_NAME
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
encoded_dataset = ds.map(lambda examples: tokenizer(examples['seq']))
```
